```python
from transformers import pipeline

classifier = pipeline('sentiment-analysis')
classifier(
    ['I am Happy', 'I feel sad']
)
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    ["We are launching a new product line to capture a larger share of the market.", "This is a course about the Transformers library"],
    candidate_labels=["education", "politics", "business"],
)
```
